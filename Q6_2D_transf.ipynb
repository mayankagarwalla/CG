```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a 2D object as a set of points
c = eval(input("Enter points in form of 2D list : "))
points = np.array(c)
```

```python
# Translation transformation matrix
def translate(dx, dy):
    return np.array([[1, 0, dx],
                     [0, 1, dy],
                     [0, 0, 1]])
```

```python
# Scaling transformation matrix
def scale(sx, sy):
    return np.array([[sx, 0, 0],
                     [0, sy, 0],
                     [0, 0, 1]])
```

```python
# Rotation transformation matrix
def rotate(theta):
    return np.array([[np.cos(theta), -np.sin(theta), 0],
                     [np.sin(theta), np.cos(theta), 0],
                     [0, 0, 1]])
```

```python
# Apply the transformation matrix to the set of points
def transform(points, T):
    homogeneous_points = np.ones((points.shape[0], points.shape[1]+1))
    homogeneous_points[:, :-1] = points
    transformed_points = np.dot(homogeneous_points, T.T)
    return transformed_points[:, :-1]
```

```python
# Plot the original and transformed object
def plot_object(points, transformed_points):
    plt.plot(points[:,0], points[:,1], 'b', label='Original')
    plt.plot(transformed_points[:,0], transformed_points[:,1], 'r', label='Transformed')
    plt.legend()
    plt.show()
```

```python
# Translation
dx, dy = 1, 2
T = translate(dx, dy)
transformed_points = transform(points, T)
plot_object(points, transformed_points)
```

```python
# Scaling
sx, sy = 2, 2
T = scale(sx, sy)
transformed_points = transform(points, T)
plot_object(points, transformed_points)
```

```python
# Rotation
theta = np.pi/4
T = rotate(theta)
transformed_points = transform(points, T)
plot_object(points, transformed_points)
```

```python
#REFLECTION

# Define the vertices of the original triangle
triangle = np.array([[0, 0], [0, 2], [2, 0], [0,0]])

# Define the reflection axis
reflection_axis = np.array([[1, 0], [0, -1]])

# Reflect the triangle by multiplying each vertex by the reflection axis
new_triangle = np.dot(triangle, reflection_axis)

# Plot the original and reflected triangles
plt.plot(triangle[:,0], triangle[:,1], 'bo-', label='Original Triangle')
plt.plot(new_triangle[:,0], new_triangle[:,1], 'go-', label='Reflected Triangle')
plt.axis('equal')
plt.legend()
plt.show()
```

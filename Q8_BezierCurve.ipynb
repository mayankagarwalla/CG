```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bezier_curve(p0, p1, p2, t):      # taking 3 control points p0,p1 & p2
    return (1 - t) ** 2 * p0 + 2 * (1 - t) * t * p1 + t ** 2 * p2   # Quadratic Bezier curve equation
```

```python
# Control points
p0 = np.array([0, 0])
p1 = np.array([1, 2])
p2 = np.array([3, 1])
```

```python
# Number of points to generate on the curve
n_points = 100
```

```python
# Generate points on the curve
t_values = np.linspace(0, 1, n_points)
curve_points = np.array([bezier_curve(p0, p1, p2, t) for t in t_values])
```

```python
# Plot the curve and control points
plt.plot(curve_points[:,0], curve_points[:,1], label='Bezier curve')
plt.plot([p0[0], p1[0], p2[0]], [p0[1], p1[1], p2[1]], 'ro', label='Control points')
plt.legend()
plt.show()
```
